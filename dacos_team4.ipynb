```python
# loading dataset
import pandas as pd
dataset = pd.read_excel('/content/drive/MyDrive/지역 유통(판매)데이터.xlsx', sheet_name=1)
```

```python
# checking dataset
print(len(dataset))
dataset.sample(n=5)
```

```python
# 전처리
# 구분, 상품명 필터링

salesdata = dataset.loc[(dataset['구분'] == '매출')
        & (dataset['상품명'].str.contains("제로 | 닭가슴살 | 프로틴 | 그릭 | 탄산수 | 트레비 | 씨그램 | 단백 | 곤약 | 다이어트"))]
filter = dataset[dataset['상품명'].str.contains("홈런볼 | 생리대")].index
salesdata = salesdata.drop(filter)

print(len(salesdata))
salesdata.head()
```
